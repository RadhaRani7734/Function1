6. Explain the concept of generators in Python and how they are defined.
- Generators in Python are a special type of function that return an iterator object and allow you to iterate over a sequence of values lazily, meaning they generate items on the fly and do not store the entire sequence in memory. This makes generators memory-efficient especially for large datasets or infinite sequences.​

# How Generators are Defined
A generator function is defined like a normal function using the def keyword, but instead of return, it uses the yield keyword to yield values one at a time.

Each time the generator’s __next__() method is called, execution resumes after the last yield statement until the next yield is encountered or the function terminates.

```python
def count_up_to(n):
    count = 1
    while count <= n:
        yield count  # pauses here and returns count
        count += 1

for number in count_up_to(5):
    print(number)
```

# 7. What are the advantages of using generators over regular functions?
Generators in Python offer several advantages over regular functions, especially when dealing with large data or sequences:

- Memory Efficiency: Generators produce items one at a time and only when required (lazy evaluation), which means they don’t store the entire sequence in memory. This makes them very memory-efficient compared to regular functions that return a full list.​

- Performance: Since values are generated on-the-fly, generators can start producing output immediately without waiting for all of it to be computed. This leads to faster response times and better performance for large datasets or infinite sequences.​

- Simpler Code: Generators eliminate boilerplate code associated with creating iterators (like __iter__ and __next__ methods). They use yield to produce values, making the code concise and easier to read compared to class-based iterators.​

- Infinite Sequences: Generators can represent infinite sequences, as they generate values indefinitely without memory overflow. Regular functions producing lists cannot handle infinite sequences.

```python
# Regular function returning a list (memory heavy)
def square_numbers_list(n):
    result = []
    for i in range(n):
        result.append(i * i)
    return result

# Generator function yielding squares (memory efficient)
def square_numbers_gen(n):
    for i in range(n):
        yield i * i

# Using the generator
for num in square_numbers_gen(5):
    print(num)
```

# 9. Explain the purpose and usage of the `map()` function in Python.
The map() function in Python is used to apply a given function to every item of one or more iterables (like lists or tuples) and return a map object (an iterator) with the results. It allows you to process and transform all items in the iterable(s) without explicitly using a loop.

Purpose:
To apply a transformation or function to each element in an iterable.

To simplify code by avoiding explicit loops for element-wise operations.

```python
# example
# Basic usage with a function
def square(x):
    return x * x

numbers = [1, 2, 3, 4]
squared = map(square, numbers)
print(list(squared))  # Output: [1, 4, 9, 16]

# Using lambda with map
numbers = [1, 2, 3, 4]
squared = map(lambda x: x*x, numbers)
print(list(squared))  # Output: [1, 4, 9, 16]

# Using map with multiple iterables
a = [1, 2, 3]
b = [4, 5, 6]
summed = map(lambda x, y: x + y, a, b)
print(list(summed))  # Output: [5, 7, 9]
```

# 10. What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?
- The difference between map(), reduce(), and filter() functions in Python lies in their purpose and how they process data from iterables:
# map()
- Applies a given function to every item of one or more iterables
- Function + one or more iterables
- Iterator with transformed elements
# reduce()
- Filters elements from an iterable based on a function’s condition
- Function (returns bool) + one iterable
- Iterator with elements where function is True
- Select elements by condition
# filter
Applies a function cumulatively to reduce iterable to a single value
Function + one iterable
Single cumulative value
Aggregate or combine items

```python
# map() example:


numbers = [1, 2, 3, 4]
squared = list(map(lambda x: x**2, numbers))
print(squared)
#filter() example:


numbers = [1, 2, 3, 4, 5]
even_numbers = list(filter(lambda x: x % 2 == 0, numbers))
print(even_numbers)

#reduce() example:


from functools import reduce
numbers = [1, 2, 3, 4]
product = reduce(lambda x, y: x * y, numbers)
print(product)
```

# 4. Write a Python function that checks if a given number is prime or not from 1 to 200.

```python
def is_prime(num):
    if num <= 1 or num > 200:
        return False
    if num == 2:
        return True
    if num % 2 == 0:
        return False
    for i in range(3, int(num**0.5) + 1, 2):
        if num % i == 0:
            return False
    return True


for number in range(1, 201):
    if is_prime(number):
        print(f"{number} is prime")
```

# 5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms.


```python
class FibonacciIterator:
    def __init__(self, max_terms):
        self.max_terms = max_terms
        self.count = 0
        self.a, self.b = 0, 1

    def __iter__(self):
        return self

    def __next__(self):
        if self.count >= self.max_terms:
            raise StopIteration
        if self.count == 0:
            self.count += 1
            return 0
        elif self.count == 1:
            self.count += 1
            return 1
        else:
            self.a, self.b = self.b, self.a + self.b
            self.count += 1
            return self.a

# Example usage:
fib_iter = FibonacciIterator(10)
for num in fib_iter:
    print(num)
```

# 6. Write a generator function in Python that yields the powers of 2 up to a given exponent

```python
def powers_of_two(max_exponent):
    for exp in range(max_exponent + 1):
        yield 2 ** exp


for value in powers_of_two(5):
    print(value)
```

# 7. Implement a generator function that reads a file line by line and yields each line as a string.

```python
def read_file_line_by_line(file_path):
    with open(file_path, 'r') as file:
        for line in file:
            yield line.rstrip('\n')


for line in read_file_line_by_line('example.txt'):
    print(line)
```

# 9. Write a Python program that uses `map()` to convert a list of temperatures from Celsius to Fahrenheit.

```python
def celsius_to_fahrenheit(celsius):
    return (celsius * 9/5) + 32

celsius_temps = [0, 20, 37, 100]
fahrenheit_temps = list(map(celsius_to_fahrenheit, celsius_temps))

print(fahrenheit_temps)
```
